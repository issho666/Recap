```python
f = open("file.txt","r")
raw=f.read()
f.close()
infile = open('glove_model.pickle','rb')
glove_model = pickle.load(infile)
infile.close()
```

```python
stop_words = set(stopwords.words('english'))
raw = raw.translate(str.maketrans('', '', string.punctuation))
tokens = word_tokenize(raw)
tokens = [i.lower() for i in tokens if not i in stop_words]
print(tokens)
```

```python
def get_similarities(in_word):
    final_list=[]
    for i in tokens:
        try:
            final_list.append([glove_model.similarity(i,in_word),i])
        except:
            continue
    return final_list
```
